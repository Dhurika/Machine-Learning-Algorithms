```python
#preprocessing
```

```python
df = df.drop(['CUST_ID'],axis=1)
```

```python
df.columns
```

```python
from sklearn.impute import KNNImputer
```

```python
null_cols = df.columns[df.isnull().any()].tolist()
imputer = KNNImputer()
df_imp = pd.DataFrame(imputer.fit_transform(df[null_cols]), columns=null_cols)
df = df.fillna(df_imp)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df)).round(2)
```

```python
df.describe()
```

```python
import numpy as np
from sklearn.decomposition import PCA

df = np.asarray(df)
pca = PCA(n_components=2, random_state=24)
df = pca.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3,n_init=25,random_state=1234)
```

```python
km.fit(df)
```

```python
km.labels_
```

```python
km.inertia_ #WCSS value
```

```python
pca_df = pd.DataFrame(data=df, columns=['PC1', 'PC2'])
print(pca_df)
```

```python
cluster_centers = pd.DataFrame(km.cluster_centers_,columns=['PC1','PC2'])
```

```python
#Evaluate and visulize the clusters
```

```python
plt.figure(figsize=(10,3))

sns.scatterplot(data=pca_df,
                x="PC1",
                y="PC2",
                hue=km.labels_,
                s=100,
                alpha=0.8)

sns.scatterplot(data=cluster_centers,
                x="PC1",
                y="PC2",
                hue=cluster_centers.index,
                palette = "colorblind",
                s=100,
                marker='D')

for i in range(len(cluster_centers)):
    plt.text(x=cluster_centers.PC1[i],
             y=cluster_centers.PC2[i],
             s=i,
             horizontalalignment='center',
             verticalalignment='center')

```

```python
#Determine the right 'k'
```

```python
#1.wcss

wcss = []
for i in range(2,11):
    km = KMeans(n_clusters=i,n_init=25,random_state=1234)
    km.fit(pca_df)
    wcss.append(km.inertia_)
wcss = pd.Series(wcss,index=range(2,11))

plt.figure(figsize=(10,3))
sns.lineplot(y=wcss,x=wcss.index)
sns.scatterplot(y=wcss,x=wcss.index,s=154)
```

```python
#2.Average silhouette

from sklearn.metrics import silhouette_score
silh = []
for i in range(2,11):
    km = KMeans(n_clusters=i,n_init=25,random_state=1234)
    km.fit(pca_df)
    silh.append(silhouette_score(pca_df,km.labels_))
silh = pd.Series(silh,index=range(2,11))

plt.figure(figsize=(10,3))
sns.lineplot(y=silh,x=silh.index)
sns.scatterplot(y=silh,x=silh.index,s=154)
```

```python
#3.calinski harabasz score

from sklearn.metrics import calinski_harabasz_score
chs = []
for i in range(2,11):
    km = KMeans(n_clusters=i,n_init=25,random_state=1234)
    km.fit(pca_df)
    chs.append(calinski_harabasz_score(pca_df,km.labels_))
chs = pd.Series(chs,index=range(2,11))

plt.figure(figsize=(10,3))
sns.lineplot(y=chs,x=chs.index)
sns.scatterplot(y=chs,x=chs.index,s=154)
```
